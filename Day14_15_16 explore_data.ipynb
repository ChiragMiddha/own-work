```python
def age_treat(x):
    age=x['Age']
    sex=x['Sex']
    pclass=x['Pclass']
    if pd.isnull(age):
        if sex=='male' and pclass==1:
            return df['Age'][(df['Sex']=='male') & (df['Pclass']==1)].mean()
        elif sex=='male' and pclass==2:
            return df['Age'][(df['Sex']=='male') & (df['Pclass']==2)].mean()
        elif sex=='male' and pclass==3:
            return df['Age'][(df['Sex']=='female') & (df['Pclass']==3)].mean()
        elif sex=='female' and pclass==1:
            return df['Age'][(df['Sex']=='female') & (df['Pclass']==1)].mean()
        elif sex=='female' and pclass==2:
            return df['Age'][(df['Sex']=='female') & (df['Pclass']==2)].mean()
        elif sex=='female' and pclass==3:
            return df['Age'][(df['Sex']=='female') & (df['Pclass']==3)].mean()
    return age
```

```python
def age_treat(x):
    age=x['Age']
    sex=x['Sex']
    pclass=x['Pclass']
    if pd.isnull(age):
        return df['Age'][(df['Sex']==sex) & (df['Pclass']==pclass)].mean()
    return age
```
